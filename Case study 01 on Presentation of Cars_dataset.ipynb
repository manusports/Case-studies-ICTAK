### Qn. 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
cars['Buyer Gender'].value_counts()
```

```python
plt.figure(figsize=(8,4))
sns.set_theme(style='whitegrid')
sns.countplot(x='Buyer Gender',data=cars, hue='Buyer Gender')
plt.xlabel('Buyer Gender')
plt.ylabel('Cars Count')
plt.title('Male vs Female Buyers',size=16)
plt.show()
```

##### Insights Qn. 3.
    1.104 Female buyers participated for the sale more than Male buyers

#### Bar Graph with Color of car wise buyers

```python
plt.figure(figsize=(20,6))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',hue='Color',data=cars)
plt.title('Male vs Female Buyers based on color',size=16)
```

##### Insights of Color of car wise buyers
    1.All the buyers are in the range of 220 to 300 based on each color.
    2.Mauv color has maximum & Fuscia color has the least Female buyers
    3.Orange color has maximum & Green color has the least Male buyers

### Qn. 4. Find the top 5 cars based on their sales price.
