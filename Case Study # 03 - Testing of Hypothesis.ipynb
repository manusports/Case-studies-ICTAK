## Hypothesis testing; is there an increase in sales after stepping into digital marketing?

#### One tailed two sample t-test

```python
sales_before=df['Sales_before_digital_add(in $)']
sales_after=df['Sales_After_digital_add(in $)']

tvalue, pvalue=stats.ttest_rel(sales_after,sales_before,alternative='greater')
print("The t value of means is: ", round(tvalue,6))
print("The p value based on the t score is: ", round(pvalue,6))

stats.ttest_rel(sales_after,sales_before,alternative='greater')
```

#### Inferences

###### This hypothesis test gives a p-value of 0 & t-value or t-statistics is 12.0907. Since the p-value simply tells the strength of evidence in support of a null hypothesis, less than the significance level, we reject the null hypothesis & fail to reject the alternate hypothesis.
###### Conclusion: Sales after digital marketing is higher than before digital marketing.

##  The company needs to check whether there is any dependency between the features “Region” and “Manager”.

#### Chi-Square test for Independence

```python
df3=pd.crosstab(df.Region, df.Manager, margins=True)
df3
```

```python
value = np.array([df3.iloc[0][0:3].values,
                  df3.iloc[1][0:3].values,
                  df3.iloc[2][0:3].values])
stat, p, dof, expected=chi2_contingency(value)

print(chi2_contingency(value)[0:3])
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

####  Inferences

###### p-vale 0.549399 > alpha of .05
###### Since the p-value is greater than alpha, fails to reject the null hypothesis (H0) and reject the alternate hypothesis (Ha).
###### Conclusion: “Region” and “Manager” in the data are independent and there are no dependency of these features.
