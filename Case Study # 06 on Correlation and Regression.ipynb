## Qn. No. 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
corrmatrix=df.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=0.98,vmax=1,annot=True,linewidth=0.2)
plt.title('Relationship between Add expenses and profit',fontsize=20,y=1)
```

```python
# Feature reduction

y=df['Profit(in Thousands)']
x=df.drop(['Date','Profit(in Thousands)'],axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
```

```python
# Checking for accuracy of the model using predictions and test values 

#1) Mean Square Error

from sklearn.metrics import mean_squared_error
print('MSE is:',mean_squared_error(y_test,predictions))
```

```python
#2) Coefficient of determination

from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
print("Coefficient of determination is",r2)
```

#### Insight

1.	Low Mean Squared Error or MSE indicates the independent variable has more explanatory power on the dependent variable.
2.	Coefficient of determination is 1 which indicates a perfect model.
3.	While executing feature reduction all ads contribute to a lower MSE and higher Coefficient of Determination or r^2. Hence no ads are dropped while regressing.
